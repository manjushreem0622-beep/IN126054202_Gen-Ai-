# **Task 1: Conceptual Understanding**

1. **What is the difference between "Love" and "love" in NLP?**

   In NLP, "Love" and "love" are treated as two different words because of the difference in case (uppercase and lowercase). This can increase the number of unique tokens unnecessarily.

   To avoid this, we usually convert all text to lowercase during preprocessing so that both "Love" and "love" are treated as the same word. This helps in reducing duplication and improving consistency.

2. **What happens if stopwords are not removed?**
   
   If stopwords are not removed, the text will contain many common words like "is", "the", "and", etc. These words do not add much meaning but appear very frequently. This can increase noise in the data and may affect the performance of machine learning models. However, in some cases, stopwords are important for understanding meaning, so removing them should be done carefully.


3. **Provide two real-world scenarios where removing stopwords can be harmful?**
   

   * **Sentiment Analysis:**
     In a sentence like "I do not like this product", the word "not" is very important. If it is removed, the sentence becomes "I like this product", which changes the meaning completely.
    
   * **Search Engines/ Question Answering:**
   In phrases like "to be or not to be", small words are important for the actual meaning. Removing them may distort the sentence and affect search results.



4. **Explain the difference between stemming and lemmatization?**
   
     **Stemming** is a process that removes the ending of words to get the root form. It is a simple and fast method, but sometimes it produces incomplete or incorrect words (e.g., "studies" → "studi")

     **Lemmatization**, on the other hand, converts a word into its correct base or dictionary form using proper linguistic rules (e.g., "running" → "run").
     So, stemming is faster but less accurate, while lemmatization is more accurate but slightly more complex.

# **Task 2:Build Advanced Preprocessing Function**


```python
import re

def preprocess_text(text):

    if text is None:
        return []

    text = str(text).strip()

    if text == "":
        return []

    # Convert to lowercase
    text = text.lower()

    # Remove URLs
    text = re.sub(r'https?://\S+|www\.\S+', ' ', text)

    # Remove email-like patterns
    text = re.sub(r'\S+@\S+', ' ', text)

    # Remove numbers
    text = re.sub(r'\d+', ' ', text)

    # Handle repeated characters (soooo → soo)
    text = re.sub(r'(.)\1{2,}', r'\1\1', text)

    # Remove special characters
    text = re.sub(r'[^a-z\s]', ' ', text)

    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    # Tokenize
    words = text.split()

    # Remove short words (<=2) but keep "no", "not"
    final_words = []
    for word in words:
        if len(word) > 2 or word in ["no", "not"]:
            final_words.append(word)

    return final_words
```

```python
print("EXAMPLES OUTPUT\n")

Test_cases = [
    "I have 2 dogs",
    "This   is   good",
    "soooo gooooood!!!",
    "WOW!!! This is GREAT!!!",
    "I am at home",
    "I am not happy",
    "Visit http://example.com now",
    "Contact me at hello@gmail.com"
]

for i, text in enumerate(Test_cases, 1):
    tokens = preprocess_text(text)
    clean_sentence = " ".join(tokens)

    print(f"Test_cases {i}")
    print("Original :", text)
    print("Tokens   :", tokens)
    print("Cleaned  :", clean_sentence)
    print("-" * 50)
```

# **Task 3: Stress Testing**

```python
test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Noooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

print("STRESS TESTING\n")

for i, sentence in enumerate(test_sentences, 1):

    tokens = preprocess_text(sentence)
    clean_sentence = " ".join(tokens)

    print(f"Sentence {i}")
    print("Original Text   :", sentence)
    print("Cleaned Tokens  :", tokens)
    print("Cleaned Sentence:", clean_sentence)
    print("-" * 60)
```

# **Task 4: Token Analytics**

```python
def token_analytics(tokens):

    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))

    if total_tokens == 0:
        avg_length = 0
    else:
        avg_length = round(sum(len(word) for word in tokens) / total_tokens, 2)

    return total_tokens, unique_tokens, avg_length


print("TOKEN ANALYTICS\n")

results = []

for i, sentence in enumerate(test_sentences, 1):

    tokens = preprocess_text(sentence)
    clean_sentence = " ".join(tokens)

    total, unique, avg_len = token_analytics(tokens)

    results.append({
        "sentence": sentence,
        "tokens": tokens,
        "total": total,
        "unique": unique
    })

    print(f"Sentence {i}")
    print("Original Text        :", sentence)
    print("Cleaned Tokens       :", tokens)
    print("Cleaned Sentence     :", clean_sentence)
    print("Total Tokens         :", total)
    print("Unique Tokens        :", unique)
    print("Average Token Length :", avg_len)
    print("-" * 60)
```

# **Task 5: Frequency Analysis**


```python
from collections import Counter

print("FREQUENCY ANALYSIS\n")

all_tokens = []

# Combine tokens from all sentences
for sentence in test_sentences:
    tokens = preprocess_text(sentence)
    all_tokens.extend(tokens)

# Count frequency
freq = Counter(all_tokens)

# Top 10 most frequent
top_10 = freq.most_common(10)

# Top 5 least frequent
least_5 = sorted(freq.items(), key=lambda x: x[1])[:5]


# Print results
print("Top 10 Most Frequent Words:")
for word, count in top_10:
    print(f"{word} : {count}")

print("\nTop 5 Least Frequent Words:")
for word, count in least_5:
    print(f"{word} : {count}")
```

# **Task 6: Build Full Pipeline**


```python
test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Noooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

def full_pipeline(text_list):

    tokens = []
    clean_sentences = []

    for text in text_list:

        cleaned_tokens = preprocess_text(text)
        cleaned_sentence = " ".join(cleaned_tokens)

        tokens.append(cleaned_tokens)
        clean_sentences.append(cleaned_sentence)

    return {
        "tokens": tokens,
        "clean_sentences": clean_sentences
    }

result = full_pipeline(test_sentences)

print("Tokens:")
for t in result["tokens"]:
    print(t)

print("\nClean Sentences:")
for s in result["clean_sentences"]:
    print(s)
```
